#.format()

With the format variable we can write variables within a string using brackets {} as containers for the result.

````python
# from Python 3
i = 3.543546
print('Test string print: {}'.format(i))

# newer form in Python 3.6+
print(f'Second test string print: {i}')

# More than one variable
apples = 2
oranges = 4

print('I have {} apples and you have {} oranges.'.format(apples, oranges)) # We can add multiple variables by separating them with ,
````

# if Statement

Executes if the boolean is True and skips the rules if it is False

````python
name = 'Alice'
if name == 'Alice':
  print('Hi, Alice.')

# Indentation is important
````

# elif Statements

elif is used to evaluate a second 'Statement', it is an acronym for else if.

````python
if name == 'Alice':
  print('Hi, Alice.')
elif age != 24:
  print('Hey, you are not Alice')
````

# else Statement

The else statement is only executed when all of our booleans are False.



````python
age = 24

if name == 'Alice':
  print('Hi, Alice.')
elif age != 24:
  print('Hey, you are not Alice')
else: print('Did you forgot your name')
````
