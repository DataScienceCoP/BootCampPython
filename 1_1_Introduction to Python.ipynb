## Binary Booleans

Binary booleans are operators that use booleans to evaluate two statements and define the result of the boolean. These operators are 'and', 'not' and 'or'.


```
# Binary Boolean help us to evaluate multiple statements.
# We can add a message if the name is outside a range.
# 
import random

guess_number = input('Guess a number from 1 to 5: ') # User entry
winning_number = random.randint(1, 5) # variable with the winning number

if guess_number == str(winning_number):
    print('Congratulations, you guessed right!') # Mensaje al ganador
elif int(guess_number) >= 6 or int(guess_number) <= 0:  
        print('Please read the instructions and try again.') # message for number out of range
else:print('The winning number was {}'.format(winning_number))
```

#.format()

With the format variable we can write variables within a string using brackets {} as containers for the result.

```
# from Python 3
i = 3.543546
print('Test string print: {}'.format(i))

# newer form in Python 3.6+
print(f'Second test string print: {i}')

# More than one variable
apples = 2
oranges = 4

print('I have {} apples and you have {} oranges.'.format(apples, oranges)) # We can add multiple variables by separating them with ,

```
