# while Loops

The declaration of a while loop allows us to execute code as long as the result of the declaration is True.


```python
spam = 0
while spam < 5:
    print('Hello, world.')
    spam = spam + 1
```

```python
# This program will not stop unless you stop it manually.
password = 1234 
while password == 1234:
  print('Very weak choose another')
```

```python
# Another while loops example
name = ''
while name != 'your name':
  print('Please type your name.')
  name = input()
  print('Thank you!')
```

```python
# while loops allow questionnaires.

answer = None

while answer != "A" and answer != "B" and answer != "C":
    answer = input("Choose an option [A,B,C]")

if answer == "A":
    print("Good choice")
elif answer == "B":
    print("There is a better option")
elif answer == "C":
    print("bad choice")
```

# break statement

Break statements allow you to evaluate a statement to immediately exit a loop.

```python
while True:
       print('Please type your name.')
       name = input()
       if name == 'your name':
         break
print('Thank you!')
```

# Continue statement 

The continue statement lets you skip the rest of the current iteration and it restarts the loop on the next iteration


```python
while True:
  print('Please confirm when you are done:')
  done = input('Are you done? ')
  if done != 'yes':
    continue
  if done == 'yes':
    break
print('done!')
```

# For loop


The for loop iterates through its clause with the variable i set to a successive value in the list [0, 1, 2, 3] on each iteration.

A common Python technique is to use range(len(someList)) with a for loop to iterate over the indices of a list. For example, enter the following in the interactive shell:

```python
for i in range(4):
    print(i)
```

```python
for letter in 'Morning':
  print(letter*5)
```

## Lists

A list is a data type that contains a series of ordered data. These lists act just like any other object within python which is why we can assign it to a variable.


```python
#list
spam = [] 
spam = ['cat', 'bat', 'rat', 'elephant']
```

## Indexes

Elements within a list are automatically numbered by an index, this index starts from 0 and can be used to locate objects within the list.

The indices also count in negative, reading from right to left.

```python
print(spam[3])
```

```python
supplies = ['pens', 'staplers', 'flamethrowers', 'binders']

for i in range(len(supplies)):
  print('Index ' + str(i) + ' in supplies is: ' + supplies[i])
```
